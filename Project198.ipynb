```python
while True:
  total=0;
  customer_name=input("Enter ur name ")
  while True:
    print("enter cost and quantity of items")
    cost=float(input("Enter the cost of items "))
    quantity=int(input("Enter the quantity of items "))
    total+=cost*quantity
    repeat=input("Do you want to repeat this process(Y/y/N/n)")
    if repeat=='n' or repeat=='N':
      break
  print(" "*50)
  print("Name: ",customer_name)
  print("total bill cost ",total)
  print()
  print("****** Thank u for shoppimg with us *****")
  print(" "*50)
  new_customer=input("Go to next person?(Y/y/N/n)")
  if new_customer=='n' or new_customer=='N':
    break
```
